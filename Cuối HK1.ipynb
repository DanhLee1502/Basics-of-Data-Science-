### Kiểm Tra Và Xóa Các Ô Dữ Liệu Bị Thiếu

```python
# Kiểm tra xem có giá trị bị thiếu không
df.isnull().sum().any()
```

```python
# Tổng số giá trị bị thiếu
df.isnull().sum().sum()
```

```python
# Kiểm tra xem dòng nào bị thiếu dữ liệu
df[pd.isnull(df).any(axis=1)]
```

```python
# Tổng dữ liệu bị thiếu ở mỗi cột
df.isnull().sum()
```

```python
# Thay thế giá trị bị thiếu của salary_in_usd bằng giá trị trung bình của salary_in_usd
df.salary_in_usd.fillna(df.salary_in_usd.mean(),inplace=True)
# Thay thế giá trị bị thiếu của work_year bằng Mode (Giá trị nào trong cột nhiều nhất sẽ thay bằng giá trị đó)
df.work_year.fillna(df.work_year.mode()[0],inplace=True)
# Thay thế giá trị bị thiếu của company_size bằng Mode (Giá trị nào trong cột nhiều nhất sẽ thay bằng giá trị đó)
df.company_size.fillna(df.company_size.mode()[0],inplace=True)
```

```python
# Kiểm tra lại giá trị bị thiếu 
df.isnull().sum()
```

### Biểu Đồ Thể Hiện Công Việc Năm 2023

```python
# Lọc Dữ Liệu Có Bao Nhiêu Công Việc Năm 2023
ft = df[df['work_year']==2023].groupby('job_title').size().nlargest(10).sort_values(ascending=False).reset_index(name='job_count')

# In Kết Quả
print (ft)

# Gán Biến Cho Biểu Đồ
fig, ax = plt.subplots(figsize=(12,6))

# Vẽ Biểu Đồ
ax = sns.barplot(ax=ax, data=ft, y=ft.job_title, x=ft.job_count, color ='#FFFF00')
ax.set_xlabel('Tổng số công việc',fontsize =15)
ax.set_ylabel('Công việc', fontsize = 15)
ax.set_title('Biểu Đồ Thể Hiện Số Lượng Công Việc Năm 2023', fontsize =15)
ax.plot()

```
